```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import random

from jittering import create_jittered_time_series
from time_warping import create_time_warped_data
```

```python
DATA_FOLDER = Path("../../data")
REAL_DATA_FOLDER = DATA_FOLDER / "real"
```

```python
data = pd.read_csv(REAL_DATA_FOLDER / "mitv_prep_1y.csv", index_col=False)
```

```python
data_jittered = create_jittered_time_series(data, jitter_factor=0.1, no_features=5, save_data=False)
data_warped = create_time_warped_data(data, save_data=False)
```

```python
# Plotten der 'Value'-Spalte
plt.plot(data['traffic_volume'][:100])

# Achsenbeschriftungen hinzufügen
plt.xlabel('Stunden')
plt.ylabel('traffic_volume')
plt.title('Original')

# Plot anzeigen
plt.show()
```

```python
# Plotten der 'Value'-Spalte
plt.plot(data_jittered['traffic_volume'][:100])

# Achsenbeschriftungen hinzufügen
plt.xlabel('Stunden')
plt.ylabel('traffic_volume')
plt.title('Jittered')

# Plot anzeigen
plt.show()
```

```python
# Plotten der 'Value'-Spalte
plt.plot(data_warped['traffic_volume'][:100])

# Achsenbeschriftungen hinzufügen
plt.xlabel('Stunden')
plt.ylabel('traffic_volume')
plt.title('Time Warped')

# Plot anzeigen
plt.show()
```
